#   data distribution for categorical variable  

```python
pd.concat([df['MSZoning'].value_counts()/df.shape[0]*100,df_drop_row['MSZoning'].value_counts()/df_drop_row.shape[0]*100],axis=1,keys=["MSZoning_org","MSZoning_clean"])
```

```python
#similarly for various columns we create one function
def cat_var_dist(var):
    return pd.concat([df[var].value_counts()/df.shape[0]*100,
               df_drop_row[var].value_counts()/df_drop_row.shape[0]*100],axis=1,keys=[var+"_org",var+"_clean"])
    
```

```python
cat_var_dist("MSZoning")
```
